```python
import pandas as pd
import pickle
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier

# 1. Load your dataset
df = pd.read_csv("synthetic_fraud_dataset.csv")

# 2. Separate features (X) and target (y)
X = df.drop(columns=['transaction_id', 'user_id', 'is_fraud'])
y = df['is_fraud']

# 3. Define which columns are text and which are numbers
numeric_features = ['amount', 'hour', 'device_risk_score', 'ip_risk_score']
categorical_features = ['transaction_type', 'merchant_category', 'country']

# 4. Create the preprocessor to handle text and numbers
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

# 5. Build the pipeline and train
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42))
])

model.fit(X, y)

# 6. Save the model
with open('fraud_model.pkl', 'wb') as f:
    pickle.dump(model, f)
print("Model saved as fraud_model.pkl")
```
